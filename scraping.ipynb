```python
from bs4 import BeautifulSoup
# importando request
import requests

# se define variable con url sólo por orden
url = 'https://gladiador-jim.github.io/Trabajo_Nro1/'

# obtención de la página que se va a analizar
page = requests.get(url)

print(f"Codigo: {page.status_code}")

# el contenido de la pagina se encuentra en page.content, y con esto se lee con BeautifulSoap
from bs4 import BeautifulSoup
page = BeautifulSoup(page.content, 'html.parser')
print(page.prettify())
```

```python
from bs4 import BeautifulSoup
# importando request
with open('Trabajo_Nro1/index.html',encoding='utf-8') as doc_html:
    page = BeautifulSoup(doc_html.read(), 'html.parser')


print(f'page: \n{page.prettify()}')

```

```python
# Obteniendo el titulo etiqueta h1
h1_tag = page.h1

# Viendo que tiene la variable h1_tag
print(f'h1_tag:\n{h1_tag}')

# Obteniendo la etiqueta h2
h2_tag = page.h2

# Viendo que tiene la variable h1_tag
print(f'h2_tag:\n{h2_tag}')

```

```python
#Identenficando la notica con la class ´texto_noticia'
class_seccion = page.find(class_='texto_noticia')
print(class_seccion)
```
