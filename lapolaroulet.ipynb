```python
from bs4 import BeautifulSoup
# importando request
import requests
import numpy

# se define variable con url sólo por orden
url = 'https://www.lapolar.cl/%F0%9F%94%A5-outlet-online/?start=0&sz=144'

# obtención de la página que se va a analizar
page = requests.get(url)

print(f"Codigo: {page.status_code}")

# el contenido de la pagina se encuentra en page.content, y con esto se lee con BeautifulSoap
from bs4 import BeautifulSoup
page = BeautifulSoup(page.content, 'html.parser')
print(page.prettify())
```

```python
all_products = page.find_all(class_='product-tile__wrapper')
print(len(all_products))
```

```python
 # obteniendo el nombre en una lista de elementos [ ] sin .string. Para conseguir el texto debe se debe elimim¿nar la variable all de la busqueda 
 # de elementos

for product in all_products:
    name_products = product.find_all(class_='link lp-font--barlow js-trunk')
    print(name_products)
```

```python
#Para tener una lista de los nombres, utilizamos string y a diferencia del anterior elimanamos los all
for product in all_products:
    name_products = product.find(class_='link lp-font--barlow js-trunk')
    print(name_products.string)
```

```python
##Obeteniendo la lista de precios con descuento por compras en internet

for product in all_products:
    name_products = product.find(class_='link lp-font--barlow js-trunk')
    print(name_products.string)

    internet = product.find(class_='internet')
    price_products = internet.find(class_='price-value')  
    print(int(price_products.string.replace('\n','').replace(' ', '').replace('$', '').replace('.', '')))

    normal = product.find(class_='normal')
    if normal is not None:
        price_products_normal = normal.find(class_='price-value')
        print(int(price_products_normal.string.replace('\n','').replace(' ', '').replace('$', '').replace('.', '')))
```
